```python
import os
import pandas as pd
import numpy as np

flu_csv = "flufighters/Deaths_from_Pneumonia_and_Influenza__P_I__and_all_deaths__by_state_and_region__National_Center_For_Health_Statistics_Mortality_Surveillance_System.csv"

flu_df = pd.read_csv(flu_csv)
flu_df.head(-10)
```

```python
flu_column_subset = pd.DataFrame(flu_df, columns = ["State", "season", "MMWR Year/Week", "Deaths from influenza", "Deaths from pneumonia", "All Deaths"])


flu_column_subset.head()
flu_column_subset["season"].count()
```

```python
# flu2017 = flu_column_subset[flu_column_subset['MMWR Year/Week]
flu_column_subset['MMWR Year/Week'] = flu_column_subset['MMWR Year/Week'].astype(str)
flu_column_subset['Year'] = flu_column_subset['MMWR Year/Week'].str[:4]
flu_column_subset['Week'] = flu_column_subset['MMWR Year/Week'].str[4:]

flu_final = pd.DataFrame(flu_column_subset, columns = ["State", "Deaths from influenza", "Deaths from pneumonia", "All Deaths", "Year", "Week"])
# flu_final.loc[flu_final['State'] != "NaN"]
# flu_final = flu_final[np.isfinite(flu_final['State'])]
# flu_final["State"].count()
flu_final = flu_final[flu_final.State!=None]
flu_final.head()
```
