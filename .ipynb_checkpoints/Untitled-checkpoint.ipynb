```python
# Dependencies
import pandas as pd

# Store filepath in a variable
file_one = "state_population_data.txt"

# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")


# Show just the header
file_one_df.head()
```

```python
three_field = file_one_df[["State or territory" , "Population estimate 2016" , "Percent of total U.S. population 2016"]].head(100)
three_field.head()
```

```python
three_field.to_csv("StateTerritory_population_v1.csv", index=False, header=True)
```
