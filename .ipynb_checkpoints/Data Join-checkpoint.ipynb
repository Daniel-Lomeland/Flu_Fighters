```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
weather_data_to_load = "Resources/weather_data.csv"
population_data_to_load = "Resources/State_population_dens_v1.csv"
flu_data_to_load="Resources/flu_data.csv"

# Read the weahter and state Data
weather_df = pd.read_csv(weather_data_to_load, low_memory=False)
population_df=pd.read_csv(population_data_to_load, low_memory=False)
flu_df=pd.read_csv(flu_data_to_load, low_memory=False)
```

```python
weather_df.head()
```

```python
population_df.head()
```

```python
merge_table = pd.merge(weather_df, population_df, on="State", how="left")
merge_table.head()
```

```python
merge_table['UID'] = merge_table['US State Name'] + merge_table['Month']
```

```python
merge_table.head()
```

```python
flu_df['UID'] = flu_df['State'] + flu_df['Month']

final_merged_table=pd.merge(merge_table, flu_df, on="UID", how="left")
final_merged_table.head()
```

```python
final_table=final_merged_table[["US State Name" , "State_x" , "Month_x","Avg_temp","Density (Pop./ mi2)","Population","sick death","all death"]]
final_table
```

```python
final_table.to_csv("Resources/final_data.csv", index=True, header=True)
```
