```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
weather_data_to_load = "Resources/weather_data.csv"
population_data_to_load = "Resources/State_population_dens_v1.csv"
flu_data_to_load="Resources/flu_data.csv"
gdp_data_to_load="Resources/GDP.csv"
age_to_load="Resources/age_demographic.csv"

# Read the weahter and state Data
weather_df = pd.read_csv(weather_data_to_load, low_memory=False)
population_df=pd.read_csv(population_data_to_load, low_memory=False, thousands=',')
flu_df=pd.read_csv(flu_data_to_load, low_memory=False)
gdp_df=pd.read_csv(gdp_data_to_load, low_memory=False)
age_df=pd.read_csv(age_to_load, low_memory=False)
```

```python
weather_pop_df = pd.merge(weather_df, population_df, on="State", how="left")
weather_pop_df.head()
```

```python
weather_pop_df['UID'] = weather_pop_df['US State Name'] + weather_pop_df['Month']
```

```python
weather_pop_df.head()
```

```python
flu_df['UID'] = flu_df['State'] + flu_df['Month']

weather_pop_flu_df=pd.merge(weather_pop_df, flu_df, on="UID", how="left")
weather_pop_flu_df.head()
```

```python
weather_pop_flu_gdp_df=pd.merge(weather_pop_flu_df, gdp_df, left_on="US State Name",right_on="State", how="left")
weather_pop_flu_gdp_df.head()
```

```python
weather_pop_flu_gdp_age_df=pd.merge(weather_pop_flu_gdp_df, age_df, left_on="US State Name",right_on="State", how="left")
weather_pop_flu_gdp_age_df.head()
```

```python
pd.to_numeric(weather_pop_flu_gdp_age_df["Population"])
weather_pop_flu_gdp_age_df["Sick_death_rate(in 1000 people_wkly)"]=weather_pop_flu_gdp_age_df["sick death"]/(weather_pop_flu_gdp_age_df["Population"]/1000)
weather_pop_flu_gdp_age_df["All_death_rate(in 1000 people_wkly)"]=weather_pop_flu_gdp_age_df["all death"]/(weather_pop_flu_gdp_age_df["Population"]/1000)
weather_pop_flu_gdp_age_df["sick of all death"]=weather_pop_flu_gdp_age_df["sick death"]/weather_pop_flu_gdp_age_df["all death"]
weather_pop_flu_gdp_age_df.head()
```

```python
final_table=weather_pop_flu_gdp_age_df[["US State Name" , "State_x" , "Month_x","Avg_temp","Density (Pop./ mi2)","Population","sick death","all death","Children 0-18","Adults 19-25","Adults 26-34","Adults 35-54","Adults 55-64","65+","Sick_death_rate(in 1000 people_wkly)","All_death_rate(in 1000 people_wkly)","sick of all death","GDP per capita"]]
final_table
```

```python
final_table.to_csv("Resources/final_data.csv", index=True, header=True)
```
